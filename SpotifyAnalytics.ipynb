Data Formatting and Normalization

```python
df_tracks.set_index('release_date', inplace=True)
```

```python
df_tracks.index=pd.to_datetime(df_tracks.index, format='mixed')
```

```python
df_tracks.head()
```

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.head()
```

EDA 

```python
df_numeric = df_tracks.select_dtypes(include=['number'])
corr_df = df_numeric.corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sbn.heatmap(corr_df,annot=True, fmt=".1g",vmin=-1,vmax=1,center=0, cmap="magma",linewidth=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
plt.figure(figsize=(10,6))
sbn.regplot(data= df_tracks, y= "loudness", x="energy", color="c").set(title="Loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sbn.regplot(data= df_tracks, y= "popularity", x="acousticness", color="r").set(title="Popularity vs Acousticness")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
 sbn.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of Songs Released per year')
```

```python
total_dr=df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig= sbn.barplot(x= years, y= total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
sbn.set_style(style="whitegrid")
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig= sbn.lineplot(x= years, y= total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation=60)
```
